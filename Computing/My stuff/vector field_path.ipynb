```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
# start a new figure
plt.figure()
# add grid lines
plt.grid(True)

#PLOT THE VECTOR FIELD
x = np.linspace(0,10,10)
y = np.linspace(0,10,10)
X,Y = np.meshgrid(x,y)
plt.quiver(X,Y,X*Y**2,X**2*Y)

#PLOT THE PATH
plt.plot(3*x,y)

#limit the x and y axes
plt.xlim(0,10)
plt.ylim(0,10)
```
