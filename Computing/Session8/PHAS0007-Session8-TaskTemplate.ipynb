This task is about animating trajectile motion. The user inputs the initial height of the ball with the velocity vector's length and it's angle with the x axis.

```python
import numpy as np
from vpython import sphere, color, rate, canvas, vector, curve
```

Using the user given initial speed and angle we calculate the x and y position of the ball along with the distance travelled in the x direction if the ball started from the origin.<br>
For this we use the following equations:<br>
<br>
$$ x(t) = x_0 + v_0 t \cos \theta $$

$$ y(t) = y_0 + v_0 t \sin \theta -\frac{1}{2}g t^2 $$

$$ \text{Range} = \frac{v_0^2 \sin(2\theta)}{g} $$

```python
### set up the scene
canvas(center = vector(50,20,0),range=50)
ground = curve(pos=[vector(-5,0,0),vector(100,0,0)],color = color.green)

# other parameters
g = 9.8 # gravitational acceleration, m/s2
dt = 0.01 # time interval for loop animation, in seconds
x0 = 0.0 # initial ball x-coordinate (metres)

# input initial conditions (height, angle and speed)
y0 = float(input("Input the initial height in metres: "))
h = y0
dtheta = float(input("Input the initial angle in degrees: "))
theta = np.radians(dtheta)
v0 = float(input("Input the initial speed in metres/second: "))

# calculate Vx and Vy from the user input v0
vX = v0*np.cos(theta)
vY = v0*np.sin(theta)

# start the animation
ball = sphere(pos = vector(x0,y0,0),radius = 1,make_trail=True)
t = 0 # initial time

posY = y0
posX = x0

# run the loop until the ball's position is equal to or below 0
# incrementing t 
# each time the loop is run the new positions of the ball are calculated 
while posY >= 0: 
    t+=0.05
    rate(15)
    posY = y0 + vY*t - 9.81/2*t**2
    posX = x0 + vX*t
    ball.pos = vector(posX,posY,0)
    if posY >= h:
        rng_prog = posX 
        
# calculating the distance travelled in the x direction      
rng_calc = v0**2 *np.sin(2*theta)/9.81

# printing the results
print("The ball was in the air for ",t," seconds.")
print("The ball landed at the point ({0},{1})".format(posX,0))
print("If the ball was lakunched from the origin (0,0) with the given initial velovity it would travel:")
print("Using the equation: ",rng_calc, "meters")
print("Using the program: ",rng_prog, "meters")
print("The difference between the two: ", rng_calc-rng_prog, "meters")
```
